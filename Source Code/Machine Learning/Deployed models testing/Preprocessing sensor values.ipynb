```python
sensorData = pd.read_csv("SensorValues.csv")
```

```python
print(sensorData.head)
```

```python
pH = sensorData['pH']
EC = sensorData['EC']
P = sensorData['P']
sampleID = sensorData['sampleID']
type(pH)
```

```python
preProcessID = []
for sample in sampleID:
    tempID = sample.split("_")
    tempID = tempID[0]
    preProcessID.append(tempID)
preProcessID = pd.Series(preProcessID, name="sampleID")
print(preProcessID)
```

```python
sensorData = pd.concat([pH, P, EC, preProcessID], axis=1)
print(sensorData)
```

```python
averageValues = sensorData.groupby('sampleID', as_index=False).mean()
```

```python
print(averageValues)
```

```python
averageValues.to_csv("PreprocessedSensorValues.csv")
```
